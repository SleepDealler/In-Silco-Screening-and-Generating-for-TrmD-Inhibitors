```python
import pandas as pd

filenames = ['./ipz4.csv', './ipz5.csv', './ipz6.csv', './results4.csv', './results5.csv', './results6.csv', './results7.csv']
df = pd.concat([pd.read_csv(filename, index_col=0) for filename in filenames])
```

```python
df.index = df.index.map(lambda x: x.split('\\')[0])
df
```

```python
(df.confidence > 0).sum(), (df.confidence >= -0.5).sum()
```

```python
df.sort_values(by='confidence', ascending=False).to_csv('final_results.csv')
```

```python
df = df.sort_values(by='confidence', ascending=False)
df.reset_index(inplace = True)
df.id = df.id.map(lambda x: x.split('/')[-1])
df.set_index('id', inplace=True)
# df.drop(columns=['index'], inplace=True)
# df = df[df.confidence > 0]
```

```python
all_samples = pd.read_csv('../../data/all_samples.csv')
```

```python
df_with_smiles = pd.merge(df, all_samples, left_on='id', right_on='complex_name')[['complex_name', 'confidence', 'ligand_description']]
# df_with_smiles = df_with_smiles[df_with_smiles.confidence >= -0.5]
df_with_smiles = df_with_smiles[df_with_smiles.confidence > 0]
df_with_smiles.to_csv('dataset_smiles_threshold.csv')
df_with_smiles
```
