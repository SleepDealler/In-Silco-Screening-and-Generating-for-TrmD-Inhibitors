```python
from rdkit import Chem
from rdkit.Chem import Draw
import pandas as pd
from IPython.display import display  # Import do wyświetlania w Jupyter Notebook

def display_molecule_images(smiles_file, columns=2, rows=5):
    # Wczytaj SMILES z pliku
    smiles_df = pd.read_csv(smiles_file, header=0)
    smiles_list = smiles_df["smiles"].tolist()

    # Utwórz obiekty RDKit Molecule
    molecules = [Chem.MolFromSmiles(smiles) for smiles in smiles_list]

    # Tworzenie obrazu z molekułami
    img = Draw.MolsToGridImage(
        mols=molecules[:columns * rows],  # Tylko tyle molekuł, ile mieści się w siatce
        molsPerRow=columns,
        subImgSize=(300, 300),  # Rozmiar pojedynczego obrazka molekuły
    )

    # Wyświetl obraz w notebooku
    display(img)

# Ścieżka do pliku SMILES
input_file = "./smiles.txt"

# Wyświetl obraz molekuł
display_molecule_images(input_file)
```
