```python
import pandas as pd
import os

def split_smiles_file(input_file, output_dir, smiles_column='SMILES', chunk_size=90):
    # Otwórz plik CSV jako DataFrame
    try:
        df = pd.read_csv(input_file)
    except FileNotFoundError:
        print(f"Plik {input_file} nie istnieje.")
        return

    # Upewnij się, że katalog wyjściowy istnieje
    os.makedirs(output_dir, exist_ok=True)

    # Pobierz kolumnę 'SMILES'
    smiles = df['smiles']

    # Podziel dane na partie i zapisz do plików
    for i in range(0, len(smiles), chunk_size):
        chunk = smiles[i:i + chunk_size]
        output_file = os.path.join(output_dir, f'smiles_part_{i // chunk_size + 1}.csv')
        chunk.to_csv(output_file, index=False, header=[smiles_column])
        print(f"Zapisano {output_file}")

# Przykład użycia
input_csv = 'first-iteration-final-docking-results-03.01.2025.csv'  # Zmień na ścieżkę do swojego pliku
output_directory = 'output_smiles'  # Katalog, gdzie będą zapisane pliki
split_smiles_file(input_csv, output_directory)
```

```python
import os
import pandas as pd

def combine_csv_files(folder_path, output_file):
    """
    Combines all CSV files in a folder into a single DataFrame and saves the result.

    :param folder_path: Path to the folder containing the CSV files
    :param output_file: Path to the output CSV file
    """
    try:
        # List all files in the folder
        files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

        # Ensure there are CSV files to process
        if not files:
            print("No CSV files found in the folder.")
            return

        # Initialize an empty list to hold DataFrames
        dataframes = []

        # Read each CSV file and append it to the list
        for file in files:
            file_path = os.path.join(folder_path, file)
            df = pd.read_csv(file_path)
            dataframes.append(df)

        # Concatenate all DataFrames into a single DataFrame
        combined_df = pd.concat(dataframes, ignore_index=True)

        # Save the combined DataFrame to the output file
        combined_df.to_csv(output_file, index=False)

        print(f"Successfully combined {len(files)} files into {output_file}.")

    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage:
# Replace 'your_folder_path' with the path to your folder containing CSV files
# Replace 'output_file.csv' with your desired output file path
combine_csv_files("adme", "output_file.csv")
```

```python
import os
import pandas as pd

df = pd.read_csv('output_file.csv', sep='\t')
df_water = df[df['Water solubility'] > -4]
df_caco = df_water[df_water['Caco2 permeability'] > 1]
df_intensal = df_caco[df_caco['Intestinal absorption (human)'] > 80]
df_ames = df_intensal[df_intensal['AMES toxicity'] == 'No']
df_hepatox = df_ames[df_ames['Hepatotoxicity'] == 'No']
df_ld50 = df_hepatox[df_hepatox['Oral Rat Acute Toxicity (LD50)'] > 2]
df_uppervdss = df_ld50[df_ld50['VDss (human)'] < 2]
df_lowervdss = df_uppervdss[df_uppervdss['VDss (human)'] > 0]

filtered_df = df_lowervdss[
    (df_lowervdss['CYP2D6 substrate'] == 'No') &
    (df_lowervdss['CYP3A4 substrate'] == 'No') &
    (df_lowervdss['CYP1A2 inhibitior'] == 'No') &
    (df_lowervdss['CYP2C19 inhibitior'] == 'No') &
    (df_lowervdss['CYP2C9 inhibitior'] == 'No') &
    (df_lowervdss['CYP2D6 inhibitior'] == 'No') &
    (df_lowervdss['CYP3A4 inhibitior'] == 'No')
]
filtered_df

file_name = 'final_output.csv'
filtered_df.to_csv(file_name, index=False, encoding='utf-8')
```
